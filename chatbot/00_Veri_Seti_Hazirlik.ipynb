## 📦 Adım 1: İstanbul Veri Setini Yükle

```python
import sys, os
sys.path.insert(0, os.path.join(os.getcwd(), 'data'))
from datasets import Dataset
from data.istanbul_documents import ISTANBUL_DOCS

print("📖 İstanbul belgeleri okunuyor...")
docs_list = [{"kaynak": k, "metin": v} for k, v in ISTANBUL_DOCS.items()]
dataset = Dataset.from_list(docs_list)
print(f"\n✅ Dataset oluşturuldu!")
print(f"   Toplam kategori: {len(dataset)}")
```

## 🔍 Adım 2: Veri Setini Keşfet

```python
import pandas as pd

df = pd.DataFrame(docs_list)
df["kelime_sayisi"] = df["metin"].apply(lambda x: len(x.split()))

print("📊 Kategori İstatistikleri:")
print(df[["kaynak","kelime_sayisi"]].to_string())
print(f"\nToplam kelime: {df.kelime_sayisi.sum()}")
print(f"Ortalama kelime/kategori: {df.kelime_sayisi.mean():.1f}")
```

## 📊 Adım 3: Görselleştirme

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(12, 6))
bars = ax.barh(df["kaynak"], df["kelime_sayisi"], color="#e74c3c")
for bar, val in zip(bars, df["kelime_sayisi"]):  
    ax.text(bar.get_width()+5, bar.get_y()+bar.get_height()/2, str(val), va="center")
ax.set_xlabel("Kelime Sayısı")
ax.set_title("🕌 İstanbul Veri Seti — Kategori Başına Kelime Sayısı")
plt.tight_layout()
plt.savefig("istanbul_dataset.png", dpi=120)
plt.show()
print("✅ Grafik kaydedildi!")
```
